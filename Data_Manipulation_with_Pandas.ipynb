## 1. Sorting Data frames

 Pand Data frames has two useful functions
 -Sort_values():sort pandas data frame by one or more columns
 -Sort_index():sort pandas data frame by row index
 Suppose you want to sort the dataframe by "Outlet_Establishment_Year" then you use sort values
 
    

```python
# sort by year
sorted_data=big.sort_values(by='Outlet_Establishment_Year')
# print sorted data
sorted_data[:5]
```

```python
# sort by descending order
big.sort_values(by='Outlet_Establishment_Year', ascending=False,inplace=True)
big[:5]
```

```python
# sort by index
big.sort_index(inplace=True)
big[:]
```

```python
# Sort the entire Data by sales of oulet so that outlets that had maximum sales are on top
big.sort_values(by='Item_Outlet_Sales',ascending=False)
```

## Merging Data frames
Pandas has two useful functions that can be used to merge data frames

-concat()
 -merge()

```python
# create dumy data
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                        'B':['B0','B1','B2','B3'],
                         'C':['C0','C1','C2','C3'],
                          'D':['D0','D1','D2','D3']},
                     index=[0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
#combine data frames
#option one
result = pd.concat([df1,df2,df3])
# option two
result2 = pd.concat([df1,df2,df3] ,keys=['x','y','z'])
result2
```

```python
# Get second data frame
result2.loc['y']
```

# 2. Merging using SQL joins


```python
# create dummy data
df_a = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
result =pd.concat([df1,df2], axis=1, join='inner')
```

```python
pd.merge(df_a,df_c, on='subject_id')
```

```python
# Outer Join
pd.merge(df_a,df_b ,on ='subject_id', how='outer')
```

```python
# Inner Join
pd.merge(df_b,df_a , on ='subject_id' , how='inner')
```

```python
pd.merge(df_b,df_a, on ='subject_id' , how='left')
```

```python
# access first column by index
Bm.apply(lambda x: x[0], axis=1)
```

```python
Bm["Item_MRP"][:5]
```

```python
# clip price if its greater than 200
def clip_price(price):
    if price > 200:
        price = 200
    return price
# after clipping
Bm["Item_MRP"].apply(lambda x: clip_price(x))[:5]
```

```python
# label encode city type
def label_encode(city):
    if city =='Tier 1':
        label = 0
    elif city == 'Tier 2':
        label = 1
    else:
        label = 2
    return label
 # Operate label_encode on every row of Outlet Location Type
Bm["Outlet_Location_Type"] = Bm["Outlet_Location_Type"].apply(label_encode)
        
```
