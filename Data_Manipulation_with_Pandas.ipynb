## Merging Data frames
Pandas has two useful functions that can be used to merge data frames

-concat()

```python
# create dumy data
df1 = pd.DataFrame({'A': ['A0','A1','A2','A3'],
                        'B':['B0','B1','B2','B3'],
                         'C':['C0','C1','C2','C3'],
                          'D':['D0','D1','D2','D3']},
                     index=[0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
# clip price if its greater than 200
def clip_price(price):
    if price > 200:
        price = 200
    return price
# after clipping
Bm["Item_MRP"].apply(lambda x: clip_price(x))[:5]
```

```python
# # label encode city type
def label_encode(city):
    if city =='Tier 1':
        label = 0
    elif city == 'Tier 2':
        label = 1
    else:
        label = 2
    return label
#  # Operate label_encode on every row of Outlet Location Type
Bm["Outlet_Location_Type"] = Bm["Outlet_Location_Type"].apply(label_encode)
        
```

# 4 . Aggregation

There are multiple functions that can be used to perform useful aggregations on data in pandas:
    -groupby
    -crosstab
    -pivottable

a .***What is the mean price for each item type?:groupby***

  - In the given data set, i want to find out what is the mean price for each item type?
  
  - You can use grouby() to achieve this.
  
  - The first step would be to group the data by item_Type column

```python
# group price based on item type
price_by_item = Bm.groupby('Item_Type')

# display first few rows
price_by_item.first()
```

```python
price_by_item.Item_MRP.mean()
```

- You can use grouby with multiple columns of the dataset too
- in this case if you want to group first based on the item_Type and then item_MRP you can simply pass a list of column names

```python
# groupby on multiple columns
multiple_columns=Bm.groupby(['Item_Type', 'Item_Fat_Content'])
multiple_columns.first()
```

***b.How are outlets sizes distributed based on city type?:crosstab***
    -This functions is used to get an initial "feel"(view) of the data.Here, we can validate some basic hypothesis.
    -For example, in this case, "Outlet_Location_Type" is expected to affect "Outlet_Size" significantly.This can be tested using cross-tabulation as show below:
        

```python
Bm_2=pd.read_csv("data/bigmart.csv")
```

```python
pd.crosstab(Bm_2["Outlet_Size"],Bm_2["Outlet_Location_Type"],margins=True)
```

***c . How are sales changing per year?: pivottable***
-Most of the pivot table parameter use default values , so the only mandatory paramaters you must add are data and index.
  - ***data*** (A DataFrame to be used)
  - ***Index*** is the column to be used
  - ***values*** is the column to aggregate with.

```python
# Create pivot table
pd.pivot_table(Bm_2, index=['Outlet_Establishment_Year'])

# add value to pivot table
pd.pivot_table(Bm_2, index=['Outlet_Establishment_Year'] , values="Item_Outlet_Sales")
```

- Above examples give mean sales for each year
- Multiple columns can be passed to see mean sales not just by year but also taking into account the outlet size and city type

```python
# Create pivot table
pd.pivot_table(Bm_2,index=['Outlet_Establishment_Year', 'Outlet_Location_Type', 'Outlet_Size'], values="Item_Outlet_Sales")
```

You can also perform multiple aggregations like mean,median,min,max using ***aggfunc** parameter.

```python
pd.pivot_table(Bm_2, index=['Outlet_Establishment_Year', 'Outlet_Location_Type','Outlet_Size'], values="Item_Outlet_Sales", aggfunc=[np.mean, np.median, min,max,np.std])
```
